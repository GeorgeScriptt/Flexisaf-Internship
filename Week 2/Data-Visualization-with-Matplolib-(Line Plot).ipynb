# Data Visualization with Matplotlib (Line Chart)

## Objectives
1. Read Total Profit of all months and show it using a line plot
2. Read data for Bathing Soap and face wash of all months and display using it using a subplot

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.style.use(["ggplot"])
```

## Read Data to Pandas Dataframe

```python
file = "company_sales_data.csv"
df = pd.read_csv(file)
df
```

```python
df.describe()
```

```python
df.info()
```

## Line Plot showing Total Profits of all Months

```python
# Pre-process data for visualization
df_total = df[["month_number", "total_profit"]]

# set the month column as index
df_total = df_total.set_index("month_number")
df_total.index.name = None
df_total.head()
```

```python
# Line chart
plt.figure(figsize=(12,6))
plt.plot(df_total, color="r", marker="o")
plt.xlabel("Month Number")
plt.ylabel("Profit")
plt.title("Total Monthly Profit")
plt.xticks(df_total.index)

plt.show()
```

Observation from the Line chart
1. Profit was consistent on the `3rd` and `4th` month
2. Drastic increase in profit from the `6th` to `8th` month and `9th` to `11th` month
3. Decline in profit from `8th` to `9th` month and `11th` to `12th` month. Both occassion happend in a month interval.
4. The highest profit was incured on the `11th` and `8th` month with over 400,00 and 350,000 respectively.

## Bathing Soap and Face wash

```python
# Pre-process data for visualization
df_FB= df[["month_number", "facewash","bathingsoap"]]

# set month_number as index
df_FB = df_FB.set_index("month_number")
df_FB.index.name = None
df_FB.head()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].plot(df_FB["facewash"], color="b", marker="o")
axs[0].set_xlabel("Month Number")
axs[0].set_ylabel("Sales")
axs[0].set_title("Face Wash")

axs[1].plot(df_FB["bathingsoap"], color="g", marker="s")
axs[1].set_xlabel("Month Number")
axs[1].set_ylabel("Sales")
axs[1].set_title("Bathing Soap")

plt.tight_layout()
plt.show()
```
